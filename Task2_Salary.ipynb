```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("Salary.csv")
print(data.head())
```

```python
X = data["YearsExperience"].values
Y = data["Salary"].values

# Normalize
X = X / 100.0
Y = Y / 100.0

w = np.random.randn()
b = np.random.randn()

learning_rate = 0.1
epochs = 1000


for epoch in range(epochs):

    Y_pred = w * X + b
    loss = np.mean((Y_pred - Y) ** 2)

    # Gradients
    dw = (2 / len(X)) * np.sum((Y_pred - Y) * X)
    db = (2 / len(X)) * np.sum(Y_pred - Y)

    # Gradient Descent
    w = w - learning_rate * dw
    b = b - learning_rate * db

    if epoch % 100 == 0:
        print(
            f"Epoch {epoch}, "
            f"Loss = {loss:.6f}, "
            f"w = {w:.4f}, "
            f"b = {b:.4f}"
        )

print("\nTraining Complete")
print("Final Weight =", w)
print("Final Bias =", b)

math_score = 80

prediction = w * (math_score / 100) + b

print(
    f"\nPredicted Reading Score for Math Score {math_score}:",
    round(prediction * 100, 2)
)
```
